```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(file_path):
    epochs = []
    test_mape = []
    epoch_sum_loss = []

    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('epoch ='):
                epoch = int(line.split('=')[1].strip())
                epochs.append(epoch)
            elif line.startswith('epoch sum loss ='):
                epoch_sum_loss_value = float(line.split('=')[1].strip())
                epoch_sum_loss.append(epoch_sum_loss_value)
    
    return epochs, epoch_sum_loss

def plot_data(epochs1,epoch_sum_loss1, epochs2, epoch_sum_loss2, epochs3, epoch_sum_loss3):
    plt.figure(figsize=(10, 5))
    
    plt.plot(epochs1, epoch_sum_loss1, label='Epoch Sum Loss File 1', linestyle='-',color='green')
    plt.plot(epochs2, epoch_sum_loss2, label='Epoch Sum Loss File 2', linestyle='-',color='red')
    plt.plot(epochs3, epoch_sum_loss3, label='Epoch Sum Loss File 3', linestyle='-',color='blue')

    plt.xlabel('Epochs')
    plt.ylabel('Values')
    plt.title('Test Metrics over Epochs')
    plt.ylim(0, 1)  
    plt.legend()
    plt.show()


file_path1 = '../result#/Costco/Satellite/result_1.txt'
file_path2 = '../result#/Costco/Satellite/result_2.txt'
file_path3 = '../result#/Costco/Satellite/result_3.txt'


epochs1, epoch_sum_loss1= read_data(file_path1)
epochs2, epoch_sum_loss2 = read_data(file_path2)
epochs3, epoch_sum_loss3 = read_data(file_path3)


plot_data(epochs1, epoch_sum_loss1, epochs2, epoch_sum_loss2, epochs3, epoch_sum_loss3)
```
