```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(file_path):
    epochs = []
    test_mape = []

    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('epoch ='):
                epoch = int(line.split('=')[1].strip())
                epochs.append(epoch)
            elif line.startswith('test_mape:'):
                test_mape_value = float(line.split(':')[1].strip())
                test_mape.append(test_mape_value)
    
    return epochs, test_mape

def plot_data(epochs1,test_mape1, epochs2, test_mape2):
    plt.figure(figsize=(10, 5))
    
    plt.plot(epochs1, test_mape1, label='Test MAPE File 1', linestyle='-',color='black')
    plt.plot(epochs2, test_mape2, label='Test MAPE File 2', linestyle='-',color='red')

    plt.xlabel('Epochs')
    plt.ylabel('Values')
    plt.title('Test Metrics over Epochs')
    plt.ylim(100, 800) 
    plt.legend()
    plt.show()


file_path1 = '../result/NTC/sun/result_1.txt'
file_path2 = '../result/NTC/sun/result_2.txt'

epochs1, test_mape1= read_data(file_path1)
epochs2, test_mape2 = read_data(file_path2)


plot_data(epochs1, test_mape1, epochs2, test_mape2)
```
