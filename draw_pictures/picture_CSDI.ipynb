```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(file_path):
    epochs = []
    test_rmse = []
    test_mae = []
    epoch_sum_loss = []
    crps = []
    crps_sum = []
    test_er = []
    test_mse = []
    test_rse = []
    test_mape = []
    test_mspe = []
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('epoch ='):
                epoch = int(line.split('=')[1].strip())
                epochs.append(epoch)
            elif line.startswith('loss_total:'):
                loss_total_value = float(line.split(':')[1].strip())
                epoch_sum_loss.append(loss_total_value)
            elif line.startswith('test_rmse:'):
                rmse_value = float(line.split(':')[1].strip())
                test_rmse.append(rmse_value)
            elif line.startswith('test_mae:'):
                mae_value = float(line.split(':')[1].strip())
                test_mae.append(mae_value)
            elif line.startswith('CRPS_sum:'):
                crps_sum_value = float(line.split(':')[1].strip())
                crps_sum.append(crps_sum_value)
            elif line.startswith('test er:'):
                er_value = float(line.split(':')[1].strip())
                test_er.append(er_value)
            elif line.startswith('test_mse:'):
                mse_value = float(line.split(':')[1].strip())
                test_mse.append(mse_value)
            elif line.startswith('test_rse:'):
                rse_value = float(line.split(':')[1].strip())
                test_rse.append(rse_value)
            elif line.startswith('test_mape:'):
                mape_value = float(line.split(':')[1].strip())
                test_mape.append(mape_value)
            elif line.startswith('test_mspe:'):
                mspe_value = float(line.split(':')[1].strip())
                test_mspe.append(mspe_value)

    return epochs, test_rmse, test_mae, epoch_sum_loss, crps_sum, test_er, test_mse, test_rse, test_mape, test_mspe

def plot_data(epochs1, test_rmse1, test_mae1, epoch_sum_loss1,  crps_sum1, test_er1,test_mse1,test_rse1,test_mape1,test_mspe1,epochs2, test_rmse2, test_mae2, epoch_sum_loss2,  crps_sum2,test_er2,test_mse2,test_rse2,test_mape2,test_mspe2):
    plt.figure(figsize=(10, 5))
    
    plt.plot(epochs1, test_rmse1, label='Test RMSE File 1', linestyle='-', color='red')
    plt.plot(epochs1, test_mae1, label='Test MAE File 1', linestyle='-', color='black')
    plt.plot(epochs1, epoch_sum_loss1, label='Epoch Sum Loss File 1', linestyle='-', color='green')
    plt.plot(epochs1, crps_sum1, label='CRPS Sum File 1', linestyle='-', color='purple')
    plt.plot(epochs1, test_er1, label='Test ER File 1', linestyle='-', color='orange')
    plt.plot(epochs1, test_mse1, label='Test MSE File 1', linestyle='-', color='brown')
    plt.plot(epochs1, test_rse1, label='Test RSE File 1', linestyle='-', color='pink')
    plt.plot(epochs1, test_mape1, label='Test MAPE File 1', linestyle='-', color='cyan')
    plt.plot(epochs1, test_mspe1, label='Test MSPE File 1', linestyle='-', color='yellow')

    
    plt.plot(epochs2, test_rmse2, label='Test RMSE File 2', linestyle='--', color='red')
    plt.plot(epochs2, test_mae2, label='Test MAE File 2', linestyle='--', color='black')
    plt.plot(epochs2, epoch_sum_loss2, label='Epoch Sum Loss File 2', linestyle='--', color='green')
    plt.plot(epochs2, crps_sum2, label='CRPS Sum File 2', linestyle='--', color='purple')
    plt.plot(epochs2, test_er2, label='Test ER File 2', linestyle='--', color='orange')
    plt.plot(epochs2, test_mse2, label='Test MSE File 2', linestyle='--', color='brown')
    plt.plot(epochs2, test_rse2, label='Test RSE File 2', linestyle='--', color='pink')
    plt.plot(epochs2, test_mape2, label='Test MAPE File 2', linestyle='--', color='cyan')
    plt.plot(epochs2, test_mspe2, label='Test MSPE File 2', linestyle='--', color='yellow')
    
    plt.xlabel('Epochs')
    plt.ylabel('Values')
    plt.title('Test Metrics over Epochs')
    plt.ylim(0, 2) 
    plt.legend()
    plt.show()


file_path1 = '../result#/PriSTI/PM25/result_1_1.txt'
file_path2 = '../result#/PriSTI/PM25/result_2_1.txt'

epochs1, test_rmse1, test_mae1, epoch_sum_loss1, crps_sum1 , test_er1,test_mse1,test_rse1,test_mape1,test_mspe1  = read_data(file_path1)
epochs2, test_rmse2, test_mae2, epoch_sum_loss2,  crps_sum2 , test_er2,test_mse2,test_rse2,test_mape2,test_mspe2 = read_data(file_path2)


plot_data(epochs1, test_rmse1, test_mae1, epoch_sum_loss1,  crps_sum1, test_er1,test_mse1,test_rse1,test_mape1,test_mspe1,epochs2, test_rmse2, test_mae2, epoch_sum_loss2,  crps_sum2,test_er2,test_mse2,test_rse2,test_mape2,test_mspe2)
    
```
