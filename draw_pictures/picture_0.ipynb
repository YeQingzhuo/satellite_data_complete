```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(file_path):
    epochs = []
    test_er = []
    test_mae = []
    test_mse = []
    test_rse = []
    test_rmse = []
    test_mape = []
    test_mspe = []
    epoch_sum_loss = []
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('epoch ='):
                epoch = int(line.split('=')[1].strip())
                epochs.append(epoch)
            elif line.startswith('test er:'):
                test_er_value = float(line.split(':')[1].strip())
                test_er.append(test_er_value)
            elif line.startswith('test_mae:'):
                test_mae_value = float(line.split(':')[1].strip())
                test_mae.append(test_mae_value)
            elif line.startswith('test_mse:'):
                test_mse_value = float(line.split(':')[1].strip())
                test_mse.append(test_mse_value)
            elif line.startswith('test_rse:'):
                test_rse_value = float(line.split(':')[1].strip())
                test_rse.append(test_rse_value)
            elif line.startswith('test_rmse:'):
                test_rmse_value = float(line.split(':')[1].strip())
                test_rmse.append(test_rmse_value)
            elif line.startswith('test_mape:'):
                test_mape_value = float(line.split(':')[1].strip())
                test_mape.append(test_mape_value)
            elif line.startswith('test_mspe:'):
                test_mspe_value = float(line.split(':')[1].strip())
                test_mspe.append(test_mspe_value)
            elif line.startswith('epoch sum loss ='):
                epoch_sum_loss_value = float(line.split('=')[1].strip())
                epoch_sum_loss.append(epoch_sum_loss_value)
    
    return epochs, test_er, test_mae, test_mse, test_rse, test_rmse, test_mape, test_mspe , epoch_sum_loss

def plot_data(epochs1, test_er1, test_mae1,  test_mse1, test_rse1, test_rmse1, test_mape1, test_mspe1 , epoch_sum_loss1, epochs2, test_er2, test_mae2,  test_mse2, test_rse2, test_rmse2, test_mape2, test_mspe2 , epoch_sum_loss2):
    plt.figure(figsize=(10, 5))
    
    plt.plot(epochs1, test_er1, label='Test ER File 1', linestyle='-',color='blue')
    plt.plot(epochs1, test_mae1, label='Test MAE File 1', linestyle='-',color='black')
    plt.plot(epochs1, test_mse1, label='Test MSE File 1', linestyle='-',color='yellow')
    plt.plot(epochs1, test_rse1, label='Test RSE File 1', linestyle='-',color='orange')
    plt.plot(epochs1, test_rmse1, label='Test RMSE File 1', linestyle='-', color='red')
    plt.plot(epochs1, test_mape1, label='Test MAPE File 1', linestyle='-', color='purple')
    plt.plot(epochs1, test_mspe1, label='Test MSPE File 1', linestyle='-', color='pink')
    plt.plot(epochs1, epoch_sum_loss1, label='Epoch Sum Loss File 1', linestyle='-', color='green')
    
    plt.plot(epochs2, test_er2, label='Test ER File 2', linestyle='--',color='blue')
    plt.plot(epochs2, test_mae2, label='Test MAE File 2', linestyle='--',color='black')
    plt.plot(epochs1, test_mse2, label='Test MSE File 2', linestyle='--',color='yellow')
    plt.plot(epochs1, test_rse2, label='Test RSE File 2', linestyle='--',color='orange')
    plt.plot(epochs2, test_rmse2, label='Test RMSE File 2', linestyle='--', color='red')
    plt.plot(epochs1, test_mape2, label='Test MAPE File 2', linestyle='--', color='purple')
    plt.plot(epochs1, test_mspe2, label='Test MSPE File 2', linestyle='--', color='pink')
    plt.plot(epochs2, epoch_sum_loss2, label='Epoch Sum Loss File 2', linestyle='--', color='green')
    
    plt.xlabel('Epochs')
    plt.ylabel('Values')
    plt.title('Test Metrics over Epochs')
    plt.ylim(0, 1) 
    plt.show()


file_path1 = '../result#/NTF/PM25/day=1/rate=1%/result_x.txt'
file_path2 = '../result#/NTF/PM25/day=1/rate=1%/result_*.txt'


epochs1, test_er1, test_mae1,  test_mse1, test_rse1, test_rmse1, test_mape1, test_mspe1 , epoch_sum_loss1 = read_data(file_path1)
epochs2, test_er2, test_mae2,  test_mse2, test_rse2, test_rmse2, test_mape2, test_mspe2 , epoch_sum_loss2 = read_data(file_path2)


plot_data(epochs1, test_er1, test_mae1,  test_mse1, test_rse1, test_rmse1, test_mape1, test_mspe1 , epoch_sum_loss1, epochs2, test_er2, test_mae2,  test_mse2, test_rse2, test_rmse2, test_mape2, test_mspe2 , epoch_sum_loss2)
```
